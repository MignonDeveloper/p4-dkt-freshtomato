```python
import pandas as pd
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import mean_absolute_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 1-1. Multiple histograms on the same axis

def plot_predictions(hist, kde, best_data, new_data1, new_data2, new_data3):
    sns.distplot(best_data.prediction, 
                color="blue", label="best_data",
                hist = hist, kde = kde,
                )
    sns.distplot(new_data1.prediction, 
                hist = hist, kde = kde,
                color="red", label="new_data1")

    sns.distplot(new_data2.prediction, 
                hist = hist, kde = kde,
                color="green", label="new_data2")

    sns.distplot(new_data3.prediction, 
                hist = hist, kde = kde,
                color="skyblue", label="new_data3")

    plt.legend(title="Prediction")
    plt.show()
```

```python
best_data = pd.read_csv('/opt/p4-dkt-freshtomato/output/8072_best_file.csv')
new_data1 = pd.read_csv('/opt/p4-dkt-freshtomato/output_pycaret/0603_0612/submission/피쳐21개사용_use_only_mjh_feat_and_fillna.csv')    # 0.7957
new_data2 = pd.read_csv('/opt/p4-dkt-freshtomato/output_pycaret/0603_0618/submission/피쳐33개사용_use_ara_minyong_mjh_feat_and_fillna.csv') # 0.8017
new_data3 = pd.read_csv('/opt/p4-dkt-freshtomato/output_pycaret/0603_0715/submission/피쳐33개사용_USEFEAT-mjh-ara-minyong_FILLNA-ARA.csv')  # 0.8019	
# new_data4 = pd.read_csv('/opt/p4-dkt-freshtomato/output_pycaret/0603_0747/submission/피쳐21개사용_USEFEAT-mjh-only_FILLNA-mean.csv')  # 0.7994
new_data4 = pd.read_csv('/opt/p4-dkt-freshtomato/output_pycaret/0603_0818/submission/피쳐21개사용_USEFEAT-mjh-only_NOT-FILLNA.csv')  # 0.8016

# print(all(new_data==raw_data))
# raw_data.head()
mae = mean_absolute_error(best_data.prediction,new_data1.prediction)
msle = mean_squared_log_error(best_data.prediction,new_data1.prediction)
print(f"MAE (new_data1 vs best_data) : {mae}")
print(f"MSLE (new_data1 vs best_data) : {msle}")


mae = mean_absolute_error(best_data.prediction,new_data2.prediction)
msle = mean_squared_log_error(best_data.prediction,new_data2.prediction)
print(f"MAE (new_data2 vs best_data) : {mae}")
print(f"MSLE (new_data2 vs best_data) : {msle}")


mae = mean_absolute_error(best_data.prediction,new_data3.prediction)
msle = mean_squared_log_error(best_data.prediction,new_data3.prediction)

print(f"MAE (new_data3 vs best_data) : {mae}")
print(f"MSLE (new_data3 vs best_data) : {msle}")

mae = mean_absolute_error(best_data.prediction,new_data4.prediction)
msle = mean_squared_log_error(best_data.prediction,new_data4.prediction)

print(f"MAE (new_data4 vs best_data) : {mae}")
print(f"MSLE (new_data4 vs best_data) : {msle}")
```
